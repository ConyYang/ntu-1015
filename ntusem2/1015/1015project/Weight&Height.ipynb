## Atheletics Male

```python
Atheletics_M_win= pd.DataFrame(olympicData[["NOC", "Year","Sex","Sport","Medal","Height","Weight"]]
                     [olympicData['Year']>=1984]
                     [(olympicData['Medal']=='Gold')|
                            (olympicData['Medal']=='Silver')|
                           (olympicData['Medal']=='Bronze')]
                    [olympicData['Sport'] == 'Athletics']
                    [olympicData['Sex'] == 'M'])
Atheletics_M_lose= pd.DataFrame(olympicData[["NOC", "Year","Sex","Sport","Medal","Height","Weight"]]
                     [olympicData['Year']>=1984]
                     [(olympicData['Medal']!='Gold')&
                            (olympicData['Medal']!='Silver')&
                           (olympicData['Medal']!='Bronze')]
                    [olympicData['Sport'] == 'Athletics']
                    [olympicData['Sex'] == 'M'])
```

```python
Atheletics_M_lose['Height'].describe()
```

```python
AMW_H = Atheletics_M_win['Height']
AML_H = Atheletics_M_lose['Height']

trace_AMW_H = go.Box(
    y=AMW_H,
    name = 'Atheletics Male Win',
    marker = dict(
        color = 'rgb(183,127,112)',
    )
)
trace_AML_H = go.Box(
    y=AML_H,
    name = 'Atheletics Male Lose',
    marker = dict(
        color = 'rgb(198,219,168)',
    )
)
atheletics_M_H_layout = go.Layout(
    title = "Athletics Male Height"
)

atheletics_M_H = [trace_AMW_H , trace_AML_H]
atheletics_M_H_fig = go.Figure(data=atheletics_M_H, layout=atheletics_M_H_layout)
py.iplot(atheletics_M_H_fig ,filename = "Athletics Male Height")
```
